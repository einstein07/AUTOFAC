```python
import random
print("starting...")
line = ""
x = 0
y = 0
isOv = True
x_p = []
y_p = []
for i in range(25):
    while(isOv):
        # x coordinate
        #x = round(random.uniform(-4.5, 4.5),1)
        x = round(random.uniform(-2.8, 2.8),1)
        # y coordinate
        #y = round(random.uniform(-4.5, 4.5),1)
        y = round(random.uniform(-2, 2.8),1)
        isCheck = True
        for j in range(len(x_p)):
            if (((((x_p[j]-x)**2) + ((y_p[j]-y)**2))**0.5) < 0.8):
                isCheck = False
                break
        if (isCheck == True):
            x_p.append(x)
            y_p.append(y)
            break
        
    line = line + str(x) + " " + str(y) + " 0\n"
line = line.rstrip("\n")
print (line)
f = open("robotStartPositions.txt", "w")
f.write(line)
f.close()
print ("done")
```
